```python
#load the env variable 
from dotenv import load_dotenv

load_dotenv()
```

```python
#create an api client key
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = {"role":"user","content":text}
    messages.append(user_message)

def add_assitant_message(messages, text):
    assistant_message = {"role":"assistant","content":text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0):
    params = {
        "model":model,
        "max_tokens":1000,
        "messages":messages,
        "temperature": temperature

    }
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python

#setting the temperature to the lowest - Deterministic not very creative
messages =[]
add_user_message(
    messages,
    "Generate a one sentece movie idea",

)

answer = chat(messages, temperature=0.0)

print(answer)
```

```python

#setting the temperature to the highest - Probablistic
messages =[]
add_user_message(
    messages,
    "Generate a one sentece movie idea",

)

answer = chat(messages, temperature=1.0)

print(answer)
```

Not a dramatically different idea, just increases the chances to get a different one.
